```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/201807-citibike-tripdata.csv.zip')
data.head()
```

#Questão 1

```python
tabela = pd.DataFrame()
valores_quantis = []

for i in [0.01,0.25,0.5,0.75,0.99]:
  valores_quantis.append(data['tripduration'].quantile(q = i))

tabela.insert(0,'Quantil',['1%','25%','50%','75%','99%'])
tabela.insert(1,'Valor',valores_quantis)
#exclui os valores fora do intervalo entre o quantil 1% e o quantil 99%
data.drop(data.loc[(data['tripduration'] > valores_quantis[4]) | (data['tripduration'] < valores_quantis[0]) ].index, inplace=True)
#tabelas com os quantis
tabela
```

#Questão 2

```python
#todas as bicicletas usadas pelo menos uma vez
len(data['bikeid'].unique())
```

#Questão 3

```python
#variavel que marca o dia no mes do uso
data['day_in_month'] = data['starttime'].str[5:10]

bikes_mais_usadas = []

for i in data['bikeid'].unique():
  if len(data[data['bikeid'] == i]['day_in_month'].unique()) == 31:
    bikes_mais_usadas.append(i)
#total de bicicletas usadas todos os dias no mes
len(bikes_mais_usadas)



```

#Questão 4


```python
#todas as bicicletas usadas pelo menos uma vez
usadas = data['bikeid'].unique()
uma_aparicao = []
for i in range(len(usadas)):
  #imprime os id que apareceram apenas 1 vez
  if len(data[data['bikeid'] == usadas[i]]['day_in_month'].unique()) == 1:
    uma_aparicao.append(i)
#total de bicicletas usadas apenas uma vez no mês
len(uma_aparicao)
```

#Questão 5

```python
#agrupa as bikes por id e soma os trips durations
top_10_mais_usadas = data[['tripduration','bikeid']].groupby(by = ['bikeid'], as_index = False).sum()
#ordena conforme o tempo total decrescente
top_10_mais_usadas.sort_values(by = ['tripduration'],ascending = False).head(10)
```

#Questão 6

```python
#ordena conforme o tempo total crescente
top_10_mais_usadas.sort_values(by = ['tripduration'],ascending = True).head(10)
```

#Questão 7 

```python
#agrupa as bikes por id e soma os trips durations
bikes_total_trip_duration = data[['tripduration','bikeid']].groupby(by = ['bikeid'], as_index = False).sum()
percentual_utilizacao = []
k = 0
for i in bikes_total_trip_duration['bikeid'].unique():
  total_disponivel = 24*60*60*len(data[data['bikeid'] == i]['day_in_month'].unique())
  percentual_utilizacao.append(100* bikes_total_trip_duration['tripduration'][k]/total_disponivel)
  k+=1

utilizacao = pd.DataFrame()
utilizacao.insert(0,'Percentual de Utilização', percentual_utilizacao)
utilizacao.sort_values(by = ['Percentual de Utilização'], ascending = False)
```

```python
utilizacao['Percentual de Utilização'].mean()
#output: 7.323968855980135
utilizacao['Percentual de Utilização'].median()
#output: 7.201668263090677 
utilizacao['Percentual de Utilização'].std()
#output: 2.0224021644615453
```
